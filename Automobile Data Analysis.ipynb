##    
    . "num-of-doors","normalized-loss","bore","stroke","horse-power","peak-rpm","price" these features have the missing values.
    . We have to replace the missing value by mean value for all except num-of-doors

```python
# count the unique values in num of doors
from collections import Counter
Counter(df["num-of-doors"])
```

###  
     . Now we have to replace null values with the mean of the feature for the numeric variables and for categorical variables replace null value with the maximum count

```python
miss_col = ["normalized-losses","bore","stroke","horsepower","peak-rpm","price"]
for col in miss_col:
    df[col].replace(np.nan,df[col].astype("float").mean(axis=0),inplace=True)
    
df["num-of-doors"].replace(np.nan,df["num-of-doors"].value_counts().idxmax,inplace=True)
df.head().T
```

##  Machine Learning

###  Model Developement

#### Linear Regression

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

lm = LinearRegression()
X = df[['highway-mpg']]
Y = df['price']
lm.fit(X,Y)
y_pred = lm.predict(X)
y_pred[0:5]
```

```python
print("intercept:",lm.intercept_)
print("coefficient:",lm.coef_)
```

```python
plt.scatter(Y,y_pred)
plt.title("Predicted price by linear Regression with one variable (Highway-mpg)",fontsize=15)
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.show()
```

```python
# regression Plot
plt.figure(figsize=(8,8))
sns.regplot(x='highway-mpg',y='price',data=df)
plt.title("Regression plot",fontsize=20)
#plt.ylim(0)
plt.show()

# Residual plot
plt.figure(figsize=(8,8))
sns.residplot(x='highway-mpg',y='price',data=df)
plt.title("Regression plot",fontsize=20)
#plt.ylim(0)
plt.show()
```

```python
r2_score(Y,y_pred)
```

###  Multiple Linear Regression

```python
x = df[['horsepower','curb-weight','engine-size','highway-mpg']]
y = df['price']
mlr = LinearRegression()
mlr.fit(x,y)
pred_mlr = mlr.predict(x)
pred_mlr[0:5]
```

```python
print("intercept:",mlr.intercept_)
print("coefficient:",mlr.coef_)
```

```python
plt.scatter(y,pred_mlr)
plt.title("Predicted price by Multiple Linear Regression With more than one variables",fontsize=15)
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.show()
```

```python
# Function to plot the data.
def PollyPlot(model,x,y,name):
    x_new = np.linspace(15,55,100)
    y_new= model(x_new)
    plt.plot(x,y,'.',x_new,y_new,'_')
    plt.xlabel(name)
    plt.ylabel("price of cars")
    plt.show()
```

```python
x=df["highway-mpg"]
y=df['price']
f=np.polyfit(x,y,3)
p=np.poly1d(f)
print(p)
```

```python
PollyPlot(p,x,y,'highway-mpg')
```

###  Polynoial Linear Regression
     . We can preform a polynomial transform on on multiple features.

```python
z=df[['horsepower','curb-weight','engine-size','highway-mpg']]
pf=PolynomialFeatures(degree=2)
pf
```

```python
z.shape
```

```python
z_pf=pf.fit_transform(z)
z_pf
```

```python
plr=LinearRegression()
plr.fit(z_pf,y)
pred_plr=plr.predict(z_pf)
```

```python
plt.scatter(y,pred_plr)
plt.title("Predicted price by Polynomial Linear Regression with more than one variables",fontsize=15)
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.show()
```
